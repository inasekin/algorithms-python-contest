## Задача 1. НВП (2 балла)

Дана последовательность, требуется найти её наибольшую возрастающую подпоследовательность с помощью динамического программирования.

классическая задача на dp

dp[i] - длина наибольшей возрастающей подпоследовательности, заканчивающейся на элементе i

для каждого i смотрю на все предыдущие j < i, и если nums[j] < nums[i], то можно продлить подпоследовательность

берем максимум из всех таких вариантов + 1

сложность O(n^2)

```python
def lengthOfLIS(nums):
    if not nums:
        return 0
    
    n = len(nums)
    dp = [1] * n
    
    for i in range(1, n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    
    return max(dp)

print(lengthOfLIS([10, 9, 2, 5, 3, 7, 101, 18])) #[2,3,7,101]
print(lengthOfLIS([0, 1, 0, 3, 2, 3])) #[0,1,2,3]
print(lengthOfLIS([7, 7, 7, 7, 7, 7, 7]))
```

## Задача 2. Максимум k акций (3 балла)

Дан список prices[i], где prices[i] — цена данной акции на i-й день, и целое число k.

Определите максимальную прибыль, которую вы можете получить. **Вы можете совершить не более k транзакций: то есть вы можете купить не более k раз и продать не более k раз.**

Примечание: Вы не можете совершать несколько транзакций одновременно (то есть вы должны продать акцию, прежде чем купить её снова).

buy[i][j] - максимальная прибыль после покупки j-й транзакции в день i
sell[i][j] - максимальная прибыль после продажи j-й транзакции в день i

для покупки: либо покупаем сегодня (sell[i-1][j-1] - prices[i]), либо уже купили раньше (buy[i-1][j])
для продажи: либо продаем сегодня (buy[i-1][j] + prices[i]), либо уже продали (sell[i-1][j])

O(n*k)

```python
def maxProfit(k, prices):
    if not prices or k == 0:
        return 0

    if k >= len(prices) // 2:
        profit = 0
        for i in range(1, len(prices)):
            if prices[i] > prices[i-1]:
                profit += prices[i] - prices[i-1]
        return profit

    buy = [-float('inf')] * (k + 1)
    sell = [0] * (k + 1)
    
    for price in prices:
        for j in range(k, 0, -1):
            sell[j] = max(sell[j], buy[j] + price)
            buy[j] = max(buy[j], sell[j-1] - price)
    
    return sell[k]

print(maxProfit(2, [2, 4, 1]))
print(maxProfit(2, [3, 2, 6, 5, 0, 3]))
```

## Задача 3. Акции с охлаждением (2 балла)

Дан список prices[i], где prices[i] — цена данной акции на i-й день.

Определите максимальную прибыль, которую вы можете получить. Вы можете совершить любое количество транзакций (т.е. купить одну и продать одну акцию несколько раз) со следующими ограничениями:

**После продажи акций вы не сможете купить их на следующий день (т.е. у вас будет один день ожидания).**

Примечание: Вы не можете совершать несколько транзакций одновременно (т.е. вы должны продать акцию, прежде чем купить её снова).

отслежваем три состояния

hold - у нас есть акция (купили)
sold - только что продали, на следующий день cooldown
rest - отдыхаем, можем купить

переходы:
- hold: либо держали до этого, либо купили сегодня из rest
- sold: продали то что держали
- rest: либо уже отдыхали, либо вышли из cooldown после sold

сложность O(n)

```python
def maxProfitWithCooldown(prices):
    if len(prices) <= 1:
        return 0
    
    hold = -prices[0]
    sold = 0
    rest = 0
    
    for i in range(1, len(prices)):
        prev_hold = hold
        prev_sold = sold
        prev_rest = rest
        
        hold = max(prev_hold, prev_rest - prices[i])
        sold = prev_hold + prices[i]
        rest = max(prev_rest, prev_sold)
    
    return max(sold, rest)

print(maxProfitWithCooldown([1, 2, 3, 0, 2]))
print(maxProfitWithCooldown([1]))
```

## Задача 4. Кубики (2 балла)

Лесенкой называется набор кубиков, в котором каждый горизонтальный слой содержит меньше кубиков, чем слой под ним. Подсчитать количество различных лесенок, которые могут быть построены из N кубиков.

задача на разбиение числа

dp[n][k] - количество способов построить лесенку из n кубиков, где верхний слой содержит не более k кубиков

рекуррентность: dp[n][k] = dp[n][k-1] + dp[n-k][k-1]
- либо не используем k кубиков в верхнем слое
- либо используем k кубиков, и остается n-k кубиков с максимумом k-1

нужно вычесть 1 в конце (тк лесенка из 1 слоя не считается)

```python
def countStairs(n):
    dp = [[0] * (n + 1) for _ in range(n + 1)]

    for j in range(n + 1):
        dp[0][j] = 1
    
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            dp[i][j] = dp[i][j-1]

            if i >= j:
                dp[i][j] += dp[i-j][j-1]

    return dp[n][n] - 1

print(countStairs(3))
print(countStairs(4))
print(countStairs(6))
```

## Задача 5. Наибольшая общая подпоследовательность (2 балла)

Даны две строки text1 и text2. Верните длину их самой длинной общей подпоследовательности. Если общей подпоследовательности нет, верните 0.

Подпоследовательность строки — это новая строка, сгенерированная из исходной строки с удалением некоторых символов (может и без) без изменения относительного порядка оставшихся символов.

Общая подпоследовательность двух строк — это подпоследовательность, общая для обеих строк.

классический LCS

dp[i][j] - длина НОП для text1[:i] и text2[:j]

если символы совпадают: dp[i][j] = dp[i-1][j-1] + 1
иначе: dp[i][j] = max(dp[i-1][j], dp[i][j-1])

сложность O(m*n)

```python
def LCS(text1, text2):
    m, n = len(text1), len(text2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if text1[i-1] == text2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    
    return dp[m][n]

print(LCS("abcde", "ace"))
print(LCS("abc", "abc"))
print(LCS("abc", "def"))
```

## Задача 6. Океаны (3 балла)

Имеется прямоугольный остров размером m x n, граничащий как с Тихим, так и с Атлантическим океанами. Тихий океан омывает левый и верхний края острова, а Атлантический — правый и нижний.

Остров разбит на квадратные ячейки. Вам дана целочисленная матрица высотой m x n, где heights[r][c] представляет собой высоту над уровнем моря ячейки с координатами (r, c).

На острове выпадает много осадков, и дождевая вода может стекать в соседние ячейки с координатами (r, c) на север, юг, восток и запад, если высота соседней ячейки меньше или равна высоте текущей ячейки. Вода может стекать из любой ячейки, смежной с океаном, в океан.

Вернуть двумерный список координат сетки result, где result[i] = [ri, ci] означает, что дождевая вода может стекать из ячейки (ri, ci) как в Тихий, так и в Атлантический океан.

тут надо идти от океанов вверх по высоте

запускаем dfs от границ тихого океана (верх и лево) - помечаем все клетки откуда можно дойти
запускаем dfs от границ атлантического (низ и право) - помечаем все клетки

ответ - пересечение этих двух множеств

в dfs идем только в клетки с высотой >= текущей (вода стекает вниз, значит идем вверх)

```python
def pacificAtlantic(heights):
    if not heights or not heights[0]:
        return []
    
    m, n = len(heights), len(heights[0])
    
    pacific = set()
    atlantic = set()
    
    def dfs(r, c, visited):
        visited.add((r, c))
        
        for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < m and 0 <= nc < n and (nr, nc) not in visited:
                if heights[nr][nc] >= heights[r][c]:
                    dfs(nr, nc, visited)

    for i in range(m):
        dfs(i, 0, pacific)
    for j in range(n):
        dfs(0, j, pacific)

    for i in range(m):
        dfs(i, n-1, atlantic)
    for j in range(n):
        dfs(m-1, j, atlantic)

    result = []
    for i in range(m):
        for j in range(n):
            if (i, j) in pacific and (i, j) in atlantic:
                result.append([i, j])
    
    return result

heights = [[1,2,2,3,5],[3,2,3,4,4],[2,4,5,3,1],[6,7,1,4,5],[5,1,1,2,4]]
print(pacificAtlantic(heights))
```

## Задача 7. Распил брусьев (4 балла)

Вам нужно распилить деревянный брус на несколько кусков в заданных местах. Распилочная компания берет K рублей за распил одного бруска длиной K метров на две части.

Понятно, что различные способы распила приводят к различной суммарной стоимости заказа. Например, рассмотрим брус длиной 10 метров, который нужно распилить на расстоянии 2, 4 и 7 м, считая от одного конца. Это можно сделать несколькими способами. Можно распилить сначала на отметке 2 м, потом 4 и, наконец, 7 м. Это приведет к стоимости
10+8+6=24, потому что сначала длина бруса, который пилили, была 10 м, затем она стала 8 м, и, наконец, 6 м. А можно распилить иначе: сначала на отметке 4 м, затем 2, затем 7м. Это приведет к стоимости
10+4+6=20, что лучше.

Определите минимальную стоимость распила бруса на заданные части.

оптимальный порядок разрезов

dp[i][j] - минимальная стоимость распила отрезка от cuts[i] до cuts[j]

для каждого отрезка перебираем все возможные места первого разреза k между i и j
стоимость = (длина отрезка) + dp[i][k] + dp[k][j]

берем минимум по всем k

нужно добавить 0 и length в массив разрезов для удобства

```python
def minCost(length, cuts):
    cuts = [0] + sorted(cuts) + [length]
    n = len(cuts)

    dp = [[0] * n for _ in range(n)]

    for length_seg in range(2, n):
        for i in range(n - length_seg):
            j = i + length_seg
            dp[i][j] = float('inf')

            for k in range(i + 1, j):
                cost = (cuts[j] - cuts[i]) + dp[i][k] + dp[k][j]
                dp[i][j] = min(dp[i][j], cost)
    
    return dp[0][n-1]

print(minCost(10, [2, 4, 7]))
print(minCost(7, [1, 3, 4, 5]))
print(minCost(9, [5, 6, 1, 4, 2]))
```
