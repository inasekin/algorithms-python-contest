## Задача 1. Вхождения (1 балл)


Требуется определить в заданном массиве номер самого левого и самого правого элемента, равного искомому числу.

В тестовых случаях обязательно включите массивы с повторяющимися элементами

```python
def findLeftmost(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            result = mid
            right = mid - 1
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return result

def findRightmost(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            result = mid
            left = mid + 1
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return result

def findOccurrences(arr, target):
    leftIdx = findLeftmost(arr, target)
    rightIdx = findRightmost(arr, target)
    return leftIdx, rightIdx

print("Тест 1 - массив с повторениями:")
arr1 = [1, 2, 2, 2, 3, 4, 5]
target1 = 2
left, right = findOccurrences(arr1, target1)
print(f"Массив: {arr1}")
print(f"Искомое число: {target1}")
print(f"Левая позиция: {left}, Правая позиция: {right}")

print("\nТест 2 - массив с большим количеством повторений:")
arr2 = [5, 5, 5, 5, 5, 5, 5]
target2 = 5
left, right = findOccurrences(arr2, target2)
print(f"Массив: {arr2}")
print(f"Искомое число: {target2}")
print(f"Левая позиция: {left}, Правая позиция: {right}")

print("\nТест 3 - элемент не найден:")
arr3 = [1, 2, 3, 5, 6, 7]
target3 = 4
left, right = findOccurrences(arr3, target3)
print(f"Массив: {arr3}")
print(f"Искомое число: {target3}")
print(f"Левая позиция: {left}, Правая позиция: {right}")
```

## Задача 2. Дипломы (3 балла)





Когда Петя учился в школе, он часто участвовал в олимпиадах по информатике, математике и физике. Так как он был достаточно способным мальчиком и усердно учился, то на многих из этих олимпиад он получал дипломы. К окончанию школы у него накопилось n дипломов, причём, как оказалось, все они имели одинаковые размеры: w — в ширину и h — в высоту. Сейчас Петя учится в одном из лучших российских университетов и живёт в общежитии со своими одногруппниками. Он решил украсить свою комнату, повесив на одну из стен свои дипломы за школьные олимпиады. Так как к бетонной стене прикрепить дипломы достаточно трудно, то он решил купить специальную доску из пробкового дерева, чтобы прикрепить её к стене, а к ней — дипломы. Для того чтобы эта конструкция выглядела более красиво, Петя хочет, чтобы доска была квадратной и занимала как можно меньше места на стене. Каждый диплом должен быть размещён строго в прямоугольнике размером w на h. Дипломы запрещается поворачивать на 90 градусов. Прямоугольники, соответствующие различным дипломам, не должны иметь общих внутренних точек. Требуется написать программу, которая вычислит минимальный размер стороны доски, которая потребуется Пете для размещения всех своих дипломов.



На вход приходит три числа: w, h, n. Нужно вывести сторону доски

```python
def canFitDiplomas(side, w, h, n):
    diplomasPerRow = side // w
    diplomasPerCol = side // h
    return diplomasPerRow * diplomasPerCol >= n

def findMinBoardSize(w, h, n):
    left = 1
    right = max(w, h) * n
    result = right
    
    while left <= right:
        mid = (left + right) // 2
        if canFitDiplomas(mid, w, h, n):
            result = mid
            right = mid - 1
        else:
            left = mid + 1
    
    return result

print("Тест 1:")
w1, h1, n1 = 2, 3, 10
print(f"Ширина диплома: {w1}, Высота: {h1}, Количество: {n1}")
print(f"Минимальная сторона доски: {findMinBoardSize(w1, h1, n1)}")

print("\nТест 2:")
w2, h2, n2 = 5, 5, 4
print(f"Ширина диплома: {w2}, Высота: {h2}, Количество: {n2}")
print(f"Минимальная сторона доски: {findMinBoardSize(w2, h2, n2)}")

print("\nТест 3:")
w3, h3, n3 = 3, 4, 7
print(f"Ширина диплома: {w3}, Высота: {h3}, Количество: {n3}")
print(f"Минимальная сторона доски: {findMinBoardSize(w3, h3, n3)}")
```

## Задача 3. Коровы в стойла (2 балла)

Написать эффективную функцию для подсчета корня из целочисленного числа с точностью до k знака (можно больше)

```python
def sqrtWithPrecision(n, k):
    left = 0.0
    right = max(n, 1.0)
    precision = 10 ** (-k)
    
    while right - left > precision:
        mid = (left + right) / 2
        if mid * mid < n:
            left = mid
        else:
            right = mid
    
    return round((left + right) / 2, k)

print("Тест 1:")
n1, k1 = 16, 3
result1 = sqrtWithPrecision(n1, k1)
print(f"Корень из {n1} с точностью {k1} знаков: {result1}")
print(f"Проверка: {result1}^2 = {result1**2}")

print("\nТест 2:")
n2, k2 = 2, 5
result2 = sqrtWithPrecision(n2, k2)
print(f"Корень из {n2} с точностью {k2} знаков: {result2}")
print(f"Проверка: {result2}^2 = {result2**2}")

print("\nТест 3:")
n3, k3 = 100, 4
result3 = sqrtWithPrecision(n3, k3)
print(f"Корень из {n3} с точностью {k3} знаков: {result3}")
print(f"Проверка: {result3}^2 = {result3**2}")
```

## Задача 4. Кучки (2 балла)



Есть n кучек предметов, в i-ой кучке ai предметов. Все предметы пронумерованы сквозной нумерацией,
так, что в кучке с меньшим номером номера предметов меньше. Как за O(log n) отвечать на запрос: в какой
кучке находится предмет номер x?

(текстовое описание решения необязательно)

```python
"""
Для решения задачи используется бинарный поиск по префиксным суммам кучек. Сначала вычисляются префиксные суммы, где каждый элемент показывает общее количество предметов во всех предыдущих кучках. Затем для предмета номер x выполняется бинарный поиск в массиве префиксных сумм, чтобы найти кучку i, для которой выполняется условие: prefixSums[i] < x ≤ prefixSums[i+1]. Это гарантирует сложность O(log n).
"""
def findPile(piles, x):
    prefixSums = [0]
    for pile in piles:
        prefixSums.append(prefixSums[-1] + pile)
    
    left, right = 0, len(piles) - 1
    result = -1
    
    while left <= right:
        mid = (left + right) // 2
        if prefixSums[mid] < x <= prefixSums[mid + 1]:
            result = mid
            break
        elif x <= prefixSums[mid]:
            right = mid - 1
        else:
            left = mid + 1
    
    return result

print("Тест 1:")
piles1 = [3, 5, 2, 4]
x1 = 7
print(f"Размеры кучек: {piles1}")
print(f"Предмет номер {x1} находится в кучке: {findPile(piles1, x1)}")

print("\nТест 2:")
piles2 = [10, 20, 30, 40]
x2 = 35
print(f"Размеры кучек: {piles2}")
print(f"Предмет номер {x2} находится в кучке: {findPile(piles2, x2)}")

print("\nТест 3:")
piles3 = [5, 5, 5, 5, 5]
x3 = 13
print(f"Размеры кучек: {piles3}")
print(f"Предмет номер {x3} находится в кучке: {findPile(piles3, x3)}")
```
